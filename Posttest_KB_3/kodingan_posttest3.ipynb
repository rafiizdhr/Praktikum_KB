### Import Library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Import File Dataset

```python
df = pd.read_csv('Life Expectancy Data.csv')
```

```python
df.head()
```

### Visualisasi Data

#### Grafik Batang

```python
X = list(df["Year"])
Y = list(df["Alcohol"])

plt.figure(figsize=(10,10))
plt.title("Persentase Kematian Per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Alkohol")
plt.bar(X,Y, color="lightgreen")
plt.show()
```

## Penjelasan : 
   ### Pada grafik batang ini dapat disimpulkan bahwa persentase tertinggi alkohol menyebabkan kematian terjadi pada tahun 2007 yaitu sekitar 18% dan persentase terendah alkohol menyebabkan kematian terjadi pada tahun 2015 yaitu sekitar 11% dibandingkan dengan penyebab kematian yang lain dan rata-rata tingkat persentase kematian disebabkan oleh alkohol dari tahun 2002-2015 adalah sekitar 16%

#### Grafik Pie

```python
data_pie = df["Year"].value_counts().rename_axis("Year").reset_index(name="ID_count")
```

```python
data_pie.head()
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    data_pie["ID_count"], 
    labels=data_pie["Year"],
    autopct='%.2f%%',
    shadow=True)
plt.legend()
plt.title("Persentase Kematian Per Tahun")
plt.show()
```

### Penjelasan : 
#### Di grafik pie diatas dapat disimpulkan bahwa persentase kematian paling banyak yaitu 6.57% pada tahun 2013 dan rata-rata kematian tiap tahu adalah 6.23%

#### Grafik Scatter Plot

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='Year',y='Alcohol')
plt.title("Persentase Kematian Per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Alkohol")
plt.show()
```

### Penjelasan : 
#### Pada grafik scatter plot ini dapat disimpulkan bahwa persentase tertinggi alkohol menyebabkan kematian terjadi pada tahun 2007 yaitu sekitar 18% dan persentase terendah alkohol menyebabkan kematian terjadi pada tahun 2015 yaitu sekitar 11% dibandingkan dengan penyebab kematian yang lain dan rata-rata tingkat persentase kematian disebabkan oleh alkohol dari tahun 2002-2015 adalah sekitar 16% karena
